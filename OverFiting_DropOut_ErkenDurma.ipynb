```python
y=dataFrame["Type"].values
x=dataFrame.drop("Type",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
from keras.callbacks import EarlyStopping
```

```python
x_train.shape 
```

```python
# kural --> kaç tane kolon varsa o kadar units (gelenekleşmiş)
model=Sequential()
model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))
#sınıflandırma olduğu için 0 ve 1 var bu yüzden mse yerine binary_crossentropy
model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1)
```

```python
model.history.history
```

```python
modelKaybı=pd.DataFrame(model.history.history)
```

```python
modelKaybı.plot()
```

```python
# kural --> kaç tane kolon varsa o kadar units (gelenekleşmiş)
model=Sequential()
model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))
#sınıflandırma olduğu için 0 ve 1 var bu yüzden mse yerine binary_crossentropy
model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
earlyStopping=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
modelKaybı=pd.DataFrame(model.history.history)
```

```python
modelKaybı.plot()
```
